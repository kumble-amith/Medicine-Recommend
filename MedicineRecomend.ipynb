```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
X['Combined'] = X['Combined'].apply(stem)
X['Combined']
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
class MedicineRecomend:
    def __init__(self) -> None:
        pass
    def fit(self , X , y , medLabel,data):
        self.X = X
        self.y = y
        self.medLabel = medLabel
        self.data = data
    def recomend(self , disease ,symptom , age):
        toFind = disease + " "+symptom + " "+str(age)
        self.X = np.append(self.X , toFind)
        # print(len(self.X))
        cv = CountVectorizer(max_features=133,stop_words='english')
        vectors = cv.fit_transform(self.X).toarray()
        similar = cosine_similarity(vectors)
        dist = similar[-1]
        new_list = sorted(list(enumerate(dist)),reverse =True ,key = lambda x :x[1])
        recomend = set()
        result = []
        new_list.pop()
        for i in new_list:
            if i[0] >= len(self.data):
                continue
            med = self.data.iloc[i[0]].Medicine.strip()
            if  med not in recomend:
                med_id = self.data.iloc[i[0]].Medicine_Label.strip()
                curr = {
                    "Med_ID" :med_id,
                    "MED_Name" : med
                }
                result.append(curr)
                recomend.add(med)
            if len(recomend) == 5:
                break
        self.X = self.X[:len(self.X) -1]
        return result
```

```python
obj = MedicineRecomend()
```

```python
obj.fit(np.array(X['Combined']) , y , data['Medicine_Label'],data)
```
